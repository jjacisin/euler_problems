```python
def sum_recursive(current_number,accumulated_sum,limit):
    #Base Case
    #Return the final state
    if current_number > limit:
        return accumulated_sum
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(current_number)
        return sum_recursive(current_number+1, accumulated_sum + current_number, limit)
    
```

```python
def fib_generator(n1,n2,total,limit):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return n2
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_generator(n2, n1+n2, total,limit)
    
```

```python
def fib_summer(n1,n2,limit,total=0):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return total
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_summer(n2,n1+n2,limit,total+n1+n2)
```

```python
def fib_even_summer(n1,n2,limit,total=0):
    n3 = n1+n2
    #Base Case
    #Return the final state
    if n3 > limit:
        return total
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        if n3 % 2 == 0:
            print(n3)
            return fib_even_summer(n2,n3,limit,total+n3)
        else:
            return fib_even_summer(n2,n3,limit,total)
```

```python
# takes either a number or a number and a list of primes; latter is to speed up prime search as known primes grow
# otherwise would always check numbers in sequential (non-prime) order; wasting time
def check_prime(x,primes=[]):
    isPrime = True
    if primes:
        for num in primes:
            if x % num  == 0:
                isPrime = False

        if isPrime:
            return x

    else:
        for num in range(2,x):
            if x % num  == 0:
                isPrime = False

        if isPrime:
            return x
```

```python
def greatest_product(number,calc_length=13):
    number = str(number)
    top_score = 0
    top_string = 0
    for x in range(0,len(number)-calc_length+1):
        upper = x + calc_length
        local_str = number[x:upper]
        output = product_str(local_str)
        if output > top_score:
            top_score = output
            top_string = local_str

    return top_string, top_score
```

```python
def problem_9_orig(target_sum=1000,upper_value=750):
    triplets = []

    for x in range(1,upper_value):
        for i in range(1,x):
            a_sq = i**2
            b_sq = x**2
            c_sq = a_sq+b_sq
            c = c_sq**.5
            if (c).is_integer():
                triplets.append([i, x, c])
    
    return int(pyth_sum_checker(triplets,target_sum))
```

```python
def problem_9_refactored(target_sum=1000,upper_value=750):
    for b in range(1,upper_value):
        for a in range(1,b):
            a_sq = a**2
            b_sq = b**2
            c_sq = a_sq+b_sq
            c = c_sq**.5
            if (c).is_integer():
                if a+b+c == target_sum:
                    return int(a*b*c),[a,b,int(c)]

    return "No Pythagorean Triplets Equal Target Sum. Please check target sum or increase upper_value"
    
```
