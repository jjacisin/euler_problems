```python
def sum_recursive(current_number,accumulated_sum,limit):
    #Base Case
    #Return the final state
    if current_number > limit:
        return accumulated_sum
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(current_number)
        return sum_recursive(current_number+1, accumulated_sum + current_number, limit)
    
```

```python
sum_recursive(0,1,10)
```

```python
def fib_generator(n1,n2,total,limit):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return n2
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_generator(n2, n1+n2, total,limit)
    
```

```python
fib_generator(0,1,0,100)
```

```python
def fib_summer(n1,n2,limit,total=0):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return total
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_summer(n2,n1+n2,limit,total+n1+n2)
```

```python
fib_summer(0,1,100,0)
```

```python
def fib_even_summer(n1,n2,limit,total=0):
    n3 = n1+n2
    #Base Case
    #Return the final state
    if n3 > limit:
        return total
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        if n3 % 2 == 0:
            print(n3)
            return fib_even_summer(n2,n3,limit,total+n3)
        else:
            return fib_even_summer(n2,n3,limit,total)
```

```python
def find_largest_palindrome(lower,upper):
    pals = []
    for n1 in sorted(list(range(lower,upper)),reverse=True):
        for n2 in sorted(list(range(lower,upper)),reverse=True):
            if pal_checker(n1*n2):
                pals.append(n1*n2)
    return max(pals)
```

```python
# The sum of the squares of the first ten natural numbers is,

# 12 + 22 + ... + 102 = 385
# The square of the sum of the first ten natural numbers is,

# (1 + 2 + ... + 10)2 = 552 = 3025
# Hence the difference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 − 385 = 2640.

# Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

```

```python
def s_squares(lower,upper):
    return sum(list(map(lambda x:x**2,list(range(lower,upper+1)))))
s_squares(0,10)
```

```python
def square_s(lower,upper):
    return sum(list(range(lower,upper+1)))**2

square_s(0,10)
```

```python
def diff_square_s_s_squares(lower,upper):
    return square_s(lower,upper)-s_squares(lower,upper)

diff_square_s_s_squares(0,100)
```
