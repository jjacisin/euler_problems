```python
def fib_generator(n1,n2,total,limit):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return n2
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_generator(n2, n1+n2, total,limit)
    
```

```python
def fib_summer(n1,n2,limit,total=0):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return total
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_summer(n2,n1+n2,limit,total+n1+n2)
```

```python
# takes either a number or a number and a list of primes; latter is to speed up prime search as known primes grow
# otherwise would always check numbers in sequential (non-prime) order; wasting time
def check_prime(x,primes=[]):
    isPrime = True
    if primes:
        for num in primes:
            if x % num  == 0:
                isPrime = False
        if isPrime:
            return x

    else:
        for num in range(2,x):
            if x % num  == 0:
                isPrime = False

        if isPrime:
            return x
```

```python
def nth_prime_finder_update(n,upper=10**6):
    primes = []
    for x in range(2,upper):
        checker = check_prime(x,primes)
        if checker:
            primes.append(checker)
            if len(primes) == n:
                return primes[-1]#, primes
    return "'n' not found. The upper limit is currently set to {}. Please increase.".format(upper)
```

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
def factorial(x):
    if x == 1:
        return 1
    else:
        return x*factorial(x-1)
```

```python
def collatz_function(n,counter=0):

    #End Case
    if n == 1:
        counter += 1
        return counter
    
    else:
        if n % 2 == 0:
            n = n/2
            counter += 1
#             print('even',n)
            return collatz_function(n,counter)
        else:
            n = (3*n)+1
            counter += 1
#             print('odd',n)
            return collatz_function(n,counter)
```

## Problem 23

A perfect number is a number for which the sum of its proper divisors is exactly equal to the number. For example, the sum of the proper divisors of 28 would be 1 + 2 + 4 + 7 + 14 = 28, which means that 28 is a perfect number.

A number n is called deficient if the sum of its proper divisors is less than n and it is called abundant if this sum exceeds n.

As 12 is the smallest abundant number, 1 + 2 + 3 + 4 + 6 = 16, the smallest number that can be written as the sum of two abundant numbers is 24. By mathematical analysis, it can be shown that all integers greater than 28123 can be written as the sum of two abundant numbers. However, this upper limit cannot be reduced any further by analysis even though it is known that the greatest number that cannot be expressed as the sum of two abundant numbers is less than this limit.

Find the sum of all the positive integers which cannot be written as the sum of two abundant numbers.

```python
# Perfect Number == Divisors for N add up to N
# Abundant Number == Divisors for N add up to >N
# Deficient Number == Divisors for N add up to <N
```

```python
def find_factors(n):
    local_factors = []        
    for factor in range(1,n+1):
        if n % factor == 0:

            if local_factors:
                if factor >= local_factors[-1]:
                    break

            result = int(n / factor)
            local_factors.append(factor)
            local_factors.append(result)
    return sorted(set(local_factors))
```

```python
def problem_23_v1(upper):
    abundant = []
    deficient = []
    perfect = []
    
    for n in range(1,upper+1):
        num = find_factors(n)[:-1]
        if sum(num) > n:
            abundant.append(n)
        elif sum(num) < n:
            deficient.append(n)
        else:
            perfect.append(n)
            
    return abundant, deficient, perfect        
```

```python
def number_list_generator(upper, mode='abundant'):
    abundant = []
    deficient = []
    perfect = []

    if mode.lower() == 'abundant':
        for n in range(1,upper+1):
        
            num = find_factors(n)[:-1]
            if sum(num) > n:
                abundant.append(n)

        return abundant
    
    elif mode.lower() == 'deficient':        
        for n in range(1,upper+1):            
            num = find_factors(n)[:-1]
            if sum(num) < n:
                deficient.append(n)
        
        return deficient
    elif mode.lower() == 'perfect':
        for n in range(1,upper+1):
            
            num = find_factors(n)[:-1]
            if sum(num) == n:
                perfect.append(n)                
                
        return perfect    

    else:
        print("INCORRECT MODE GIVEN")
```
