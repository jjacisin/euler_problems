```python
def fib_generator(n1,n2,total,limit):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return n2
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_generator(n2, n1+n2, total,limit)
    
```

```python
def fib_summer(n1,n2,limit,total=0):
    #Base Case
    #Return the final state
    if n1+n2 > limit:
        return total
    
    #Recursive Case
    #Thread the state through the recursive call
    else:
        print(n1+n2)
        return fib_summer(n2,n1+n2,limit,total+n1+n2)
```

Let us list the factors of the first seven triangle numbers:

 1: 1

```python
# takes either a number or a number and a list of primes; latter is to speed up prime search as known primes grow
# otherwise would always check numbers in sequential (non-prime) order; wasting time
def check_prime(x,primes=[]):
    isPrime = True
    if primes:
        for num in primes:
            if x % num  == 0:
                isPrime = False
        if isPrime:
            return x

    else:
        for num in range(2,x):
            if x % num  == 0:
                isPrime = False

        if isPrime:
            return x
```

```python
def nth_prime_finder_update(n,upper=10**6):
    primes = []
    for x in range(2,upper):
        checker = check_prime(x,primes)
        if checker:
            primes.append(checker)
            if len(primes) == n:
                return primes[-1]#, primes
    return "'n' not found. The upper limit is currently set to {}. Please increase.".format(upper)
```

```python
def problem_12_refactored(num_of_divisors,upper=10**6):
    high_score_trip = [1]
    high_score_factors = [[1]]
    
    trip = 0

    for natural_num in range(1,upper):

        #Find current triangle num by adding natural_num to trip
        trip += natural_num


        local_factors = []        
        for factor in range(1,trip):
            if trip % factor == 0:
                
                if local_factors:
                    # KEY LINE: As 'factor' is increasing numerically, it reaches a point were 'factor' >= the previous result; then we've identified all factors
                        #For Example:
                            #triangle number = 1+2+3+4+5 = 15
                            #The loop will add 15's factors in this order: 1:15, 2:NA, 3:5, 4:NA, 5:break
                            #When the check below happens for 5 it will see that the previous result entry is 5 and will end checking any other factors
                            #As numbers become larger, this will skip a huge number of non-factors
                    if factor >= local_factors[-1]:
                        break
                
                result = int(trip / factor)
                local_factors.append(factor)
                local_factors.append(result)

                
        # Set to in case trip is a sqaure then sort
        local_factors = sorted(set(local_factors))

        
        #print high scores for progress
        if len(local_factors) > len(high_score_factors[-1]):
            high_score_trip.append(trip)
            high_score_factors.append(local_factors)
            print(len(local_factors),trip,natural_num) 

            
        #print natural_num for progress
        if natural_num % 1000 == 0:
            print(natural_num) 

            
        #End case    
        if len(local_factors) >= num_of_divisors:
            print(len(local_factors),trip)  
            return trip
```

## Problem 14

The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)

n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.

```python
def factorial(x):
    if x == 1:
        return 1
    else:
        return x*factorial(x-1)
```

```python
factorial(4)
```

```python
def collatz_function(n,counter=0):

    #End Case
    if n == 1:
        counter += 1
        return counter
    
    else:
        if n % 2 == 0:
            n = n/2
            counter += 1
#             print('even',n)
            return collatz_function(n,counter)
        else:
            n = (3*n)+1
            counter += 1
#             print('odd',n)
            return collatz_function(n,counter)
```

```python
collatz_function(13)
```

```python
def problem_14(upper = 50):
    high_score_owner = 0
    high_score = 0
    start = time.time()
    for x in range(1,upper):
        result = collatz_function(x)
        if result > high_score:
            high_score_owner = x
            high_score = result
#             if high_score % 500 == 0:
#             print(x)
    print(str(round(time.time()-start,2))+' seconds.')
    return high_score_owner, high_score
```
